```python
# notebooks/3a_repackage_downloads.ipynb

import os
import sys
from datetime import datetime
sys.path.append(os.path.abspath('../src'))

# 👇 새로 만든 '재고 조사' 부품과 기존 '패키징' 부품을 가져옵니다.
from collectors.manifest_builder import build_fileinfo_from_downloads
from collectors.package_builder import create_package

# 1. '재고 조사'를 실행하여 파일 목록을 복원합니다.
all_collected_files = build_fileinfo_from_downloads()

# 2. 복원된 파일 목록으로 패키지를 생성합니다.
if all_collected_files:
    package_name = f"data_package_repackaged_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
    create_package(package_name, all_collected_files)
else:
    print("\n⚠️ 재포장할 파일이 없습니다. data/2_downloaded/ 폴더를 확인해주세요.")
```

```python
# notebooks/tests/verify_url_generation.ipynb

import os
import sys
# src 폴더를 파이썬 경로에 추가
sys.path.append(os.path.abspath('../../src'))

# AEMCollector '부품'을 가져옵니다.
from collectors.aem_collector import AEMCollector

# --- 테스트 설정 ---
# AEMCollector를 초기화할 때는 사용자 이름과 비밀번호가 필요하지만,
# URL 생성 함수만 테스트할 것이므로 가짜 값을 넣어도 괜찮습니다.
dummy_collector = AEMCollector(username="test", password="test")

# 테스트하고 싶은 페이지 경로를 여기에 입력하세요.
test_page_path = "/content/illumina-marketing/en/areas-of-interest/cancer/clinical-cancer-research/germline-mutations"

print(f"📄 페이지 경로: '{test_page_path}'\n")
print("--- 생성된 다운로드 URL ---")

# 모든 버전에 대해 URL이 어떻게 생성되는지 확인합니다.
versions_to_test = ["sot-en", "lm-en", "lm-ko", "spac-ko_KR", "apac-en", "apac-ja_JP"]
for version in versions_to_test:
    try:
        generated_url = dummy_collector._build_aem_url(test_page_path, version)
        print(f"  - {version.ljust(12)}: {generated_url}")
    except ValueError as e:
        print(f"  - {version.ljust(12)}: 오류 - {e}")
```
